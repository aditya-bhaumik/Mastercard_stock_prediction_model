**Step 3: Preprocess the Data**



For LSTM, we need to scale the data and reshape it appropriately. For SVM and KNN, we need to handle scaling as well.

```python
from sklearn.preprocessing import MinMaxScaler
import numpy as np
```

```python
# Convert the date column to datetime
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)

# Use only the 'Close' column for simplicity
data = df['Close'].values.reshape(-1, 1)
```

```python
# Scale the data
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(data)
```

```python
# Split data into training and testing sets (80% training, 20% testing)
train_size = int(len(scaled_data) * 0.8)
test_size = len(scaled_data) - train_size
train_data, test_data = scaled_data[:train_size], scaled_data[train_size:]
```

```python
# Create a function to create a dataset with look_back
def create_dataset(dataset, look_back=1):
    X, Y = [], []
    for i in range(len(dataset)-look_back-1):
        a = dataset[i:(i+look_back), 0]
        X.append(a)
        Y.append(dataset[i + look_back, 0])
    return np.array(X), np.array(Y)
```

```python
# Reshape data for LSTM [samples, time steps, features]
look_back = 60
X_train, y_train = create_dataset(train_data, look_back)
X_test, y_test = create_dataset(test_data, look_back)
X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 1))
X_test = np.reshape(X_test, (X_test.shape[0], X_test.shape[1], 1))
```

**Step 4: Train the LSTM Model**


```python
from keras.models import Sequential
from keras.layers import Dense, LSTM
```

```python
# Create the LSTM model
lstm_model = Sequential()
lstm_model.add(LSTM(units=50, return_sequences=True, input_shape=(X_train.shape[1], 1)))
lstm_model.add(LSTM(units=50, return_sequences=False))
lstm_model.add(Dense(units=25))
lstm_model.add(Dense(units=1))
```

```python
# Compile the model
lstm_model.compile(optimizer='adam', loss='mean_squared_error')
```

```python
# Train the model
lstm_model.fit(X_train, y_train, batch_size=1, epochs=1)
```

**Step 5: Evaluate the LSTM Model**


```python
# Get the predicted values
lstm_predictions = lstm_model.predict(X_test)
lstm_predictions = scaler.inverse_transform(lstm_predictions)
```

```python
# Calculate the actual number of predictions
num_predictions = lstm_predictions.shape[0]

# Adjust the valid DataFrame to match the number of predictions
valid = df[train_size: train_size + num_predictions]

# Now assign the predictions
valid['Predictions'] = lstm_predictions
```

**Step 6: Train the SVM and KNN Models**

First, we will prepare the data for SVM and KNN by reshaping it appropriately.



```python
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error
```

```python
# Reshape data for SVM and KNN
X_train_svm, y_train_svm = create_dataset(train_data, look_back)
X_test_svm, y_test_svm = create_dataset(test_data, look_back)
```

```python
# Train the SVM model
svm_model = SVR(kernel='rbf', C=1e3, gamma=0.1)
svm_model.fit(X_train_svm, y_train_svm)
```

```python
# Train the KNN model
knn_model = KNeighborsRegressor(n_neighbors=5)
knn_model.fit(X_train_svm, y_train_svm)
```

```python
# Predict and evaluate the SVM model
svm_predictions = svm_model.predict(X_test_svm)
svm_predictions = scaler.inverse_transform(svm_predictions.reshape(-1, 1))
```

```python
# Predict and evaluate the KNN model
knn_predictions = knn_model.predict(X_test_svm)
knn_predictions = scaler.inverse_transform(knn_predictions.reshape(-1, 1))
```

**Step 7: Plotting the SVM and KNN Models**

```python
# Plot the KNN results
plt.figure(figsize=(14,7))
plt.plot(df['Close'], label='Actual Price')
plt.plot(valid.index, knn_predictions, label='KNN Predicted Price')
plt.xlabel('Date')
plt.ylabel('Close Price USD ($)')
plt.title('KNN Model')
plt.legend()
plt.show()
```

**Step 8: Calculate and print the accuracy of each model**

```python
lstm_rmse = np.sqrt(mean_squared_error(valid['Close'], lstm_predictions))
svm_rmse = np.sqrt(mean_squared_error(valid['Close'], svm_predictions))
knn_rmse = np.sqrt(mean_squared_error(valid['Close'], knn_predictions))

print(f'LSTM RMSE: {lstm_rmse}')
print(f'SVM RMSE: {svm_rmse}')
print(f'KNN RMSE: {knn_rmse}')
```
